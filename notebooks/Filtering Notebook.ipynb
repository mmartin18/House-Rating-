# Filtering notebook

```python
import geopandas as gpd
```

```python
# --- Load ORN Road Net Element layer ---
orn_path = "../data/raw/ORNELEM/ORN_ROAD_NET_ELEMENT.shp"
gdf_orn = gpd.read_file(orn_path)

# --- Filter to major roads only ---
# ORN uses ROAD_CLASS values like:
#   "Expressway", "Arterial", "Collector", "Local"
major_classes = ["Expressway", "Arterial", "Collector"]

gdf_major = gdf_orn[gdf_orn["ROAD_CLASS"].isin(major_classes)].copy()

# --- Reproject to WGS84 for consistency with your other layers ---
gdf_major = gdf_major.to_crs("EPSG:4326")

# --- Save to processed folder ---
output_path = "../data/processed/major_roads.geojson"
gdf_major.to_file(output_path, driver="GeoJSON")

print("Saved major roads to:", output_path)
```
